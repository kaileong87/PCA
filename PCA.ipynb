        resulting in 30 features.  For instance, field 0 is Mean Radius, field
        10 is Radius SE, field 20 is Worst Radius.

Features are computed from a digitized image of a fine needle
aspirate (FNA) of a breast mass.  They describe
characteristics of the cell nuclei present in the image.

Separating plane described above was obtained using
Multisurface Method-Tree (MSM-T) [K. P. Bennett, "Decision Tree
Construction Via Linear Programming." Proceedings of the 4th
Midwest Artificial Intelligence and Cognitive Science Society,
pp. 97-101, 1992], a classification method which uses linear
programming to construct a decision tree.  Relevant features
were selected using an exhaustive search in the space of 1-4
features and 1-3 separating planes.

The actual linear program used to obtain the separating plane
in the 3-dimensional space is that described in:
[K. P. Bennett and O. L. Mangasarian: "Robust Linear
Programming Discrimination of Two Linearly Inseparable Sets",
Optimization Methods and Software 1, 1992, 23-34].

First we are going to do a manual construction of PCA then followed by calling sklearn PCA library

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaled_df = scaler.fit_transform(df)
```

```python
scaled_df
```

**Steps**
1. Calculate the covariance matrix
2. Get the Eigenvectors and Eigenvalues
3. Pick some number n components
4. Sort the Eigenvalues and pick the number of components from the eigenvalue and eigenvector
5. Calculate the dot product of the original data and the eigenvectors. The output is the principal component values

```python
#step 1
covariance_matrix = np.cov(scaled_df, rowvar=False)
```

```python
covariance_matrix.shape
```

```python
#step 2
eigen_values, eigen_vector = np.linalg.eig(covariance_matrix)
```

```python
#step 3
num_components = 2
```

```python
#step 4
#In order to sort the eigenvalues and eigenvalues and pick the num_components, we need to first find the index of the sorted key
#The [::-1] flips the order, making the highest value the first.
sorted_key = np.argsort(eigen_values)[::-1][:num_components]
```

```python
eigen_values, eigen_vector = eigen_values[sort_key], eigen_vector[:, sort_key]
```

```python
#Step 5
principal_components = np.dot(scaled_df, eigen_vector)
```

```python
principal_components
```

```python
plt.figure(figsize=(12,8), dpi=200)
plt.scatter(principal_components[:,0], principal_components[:,1])
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
```

```python
cancer_dictionary['target']
```

```python
plt.figure(figsize=(12,8), dpi=200)
plt.scatter(principal_components[:,0], principal_components[:,1], c=cancer_dictionary['target'])
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
```

Now we will use sklearn to do the same

```python
from sklearn.decomposition import PCA
```

```python
model = PCA(n_components=2)
```

```python
model_PCA = model.fit_transform(scaled_df)
```

```python
plt.figure(figsize=(12,8), dpi=200)
plt.scatter(model_PCA[:,0], model_PCA[:,1], c=cancer_dictionary['target'])
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
```

```python
model.components_
```

```python
df_comp = pd.DataFrame(data=model.components_, index=['PC1', 'PC2'], columns=df.columns)
```

```python
plt.figure(figsize=(20,4), dpi=200)
sns.heatmap(df_comp, annot=True)
```

```python
model.explained_variance_ratio_
```

```python
np.sum(model.explained_variance_ratio_)
```

```python
explained_variance = []

for k in range(1,31):
    model = PCA(n_components=k)
    model.fit(scaled_df)
    
    explained_variance.append(np.sum(model.explained_variance_ratio_))
```

```python
plt.plot(range(1,31), explained_variance)
plt.xlabel('Number of components')
plt.ylabel('Variance explained')
```
